```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("customer_base.xlsx")
print(data)

years = data['Year']
saasCustomers = data['SaaS Customers']
saasRetentionRate = data['SaaS Subscription User Retention Rate']
marketingToSaas = data['Conversion(%)']
marketingCustomers = data['Marketing Customers']
marketingRetentionRate = data['Marketing Retention Rate']
```

```python
# SaaS
fig, ax1 = plt.subplots(figsize=(14, 8))

# plot SaaS customers as a bar graph
width = 0.3
customers = ax1.bar(years, saasCustomers, color='lightgrey', label='SaaS Customers', width=width)
ax1.set_ylabel('SaaS Customers', fontsize=16)
ax1.set_ylim(bottom=22000)

for bar in customers:
    height = bar.get_height()
    ax1.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=12)

# plot line graph for percentages
ax2 = ax1.twinx()
ax2.set_ylim(0, 100)
retention = sns.lineplot(x=years, y=saasRetentionRate, color='tab:red', ax=ax2, label='Retention Rate')
ax2.set_ylabel('Retention Rate (%)', fontsize=16)

# whole graph styling
ax1.set_xticks([2021, 2022, 2023])

plt.title('SaaS Customer Base and Retention', fontsize=18)
plt.show()
```

```python
# Marketing
fig, ax1 = plt.subplots(figsize=(14, 8))

# plot marketing customers as a bar graph
width = 0.3
customers = ax1.bar(years, marketingCustomers, color='lightgrey', label='Marketing Customers', width=width)
ax1.set_ylabel('Marketing Services Customers', fontsize=16)
for bar in customers:
    height = bar.get_height()
    ax1.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=12)
ax1.set_ylim(0, 1200)

# plot line graph for percentages
ax2 = ax1.twinx()
ax2.set_ylim(0, 100)
retention = sns.lineplot(x=years, y=marketingRetentionRate, color='tab:red', ax=ax2, label='Retention Rate')
ax2.set_ylabel('Retention Rate (%)', fontsize=16)

# whole graph styling
ax1.set_xticks([2021, 2022, 2023])

plt.title('Marketing Services Customer Base and Retention', fontsize=18)
plt.show()
```
